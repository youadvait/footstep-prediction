```python
import os
import librosa
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
from sklearn.model_selection import train_test_split
import warnings
warnings.filterwarnings('ignore')

# Set plotting style
plt.style.use('default')
sns.set_palette("husl")

print("‚úÖ All libraries imported successfully!")
```

```python
# Configuration
SAMPLE_RATE = 16000  # 16kHz for efficient processing
DURATION = 3.0       # 3 seconds per audio clip
RANDOM_STATE = 42    # For reproducible results

# Define paths to data folders
base_path = Path('footstepData')

# Define class folders
non_footstep_folders = [
    base_path / 'Bo6GunSounds' / 'GunReloading',
    base_path / 'Bo6GunSounds' / 'Gunshot Sounds', 
    base_path / 'UselessSoundPack'
]

footstep_folder = base_path / 'FootstepSounds'
excluded_folder = footstep_folder / 'Gun+Footsteppack'

print("üìÅ Data path configuration:")
print(f"   Base path: {base_path}")
print(f"   Sample rate: {SAMPLE_RATE} Hz")
print(f"   Duration: {DURATION} seconds")
print(f"   Excluded folder: {excluded_folder}")
```

```python
def list_audio_files(folders, exclude=None):
    """
    List all audio files in given folders, excluding specified folder
    
    Args:
        folders: List of Path objects to search
        exclude: Path object to exclude from search
    
    Returns:
        List of Path objects for found audio files
    """
    files = []
    supported_formats = ['.mp4', '.wav', '.mp3', '.m4a', '.flac']
    
    for folder in folders:
        if folder.exists():
            print(f"   üìÇ Scanning: {folder}")
            for file in folder.rglob('*'):
                if file.suffix.lower() in supported_formats:
                    # Check if file is in excluded folder
                    if exclude and exclude in file.parents:
                        print(f"   ‚è≠Ô∏è  Excluding: {file.name}")
                        continue
                    files.append(file)
                    
    return files

def check_folder_structure():
    """Check if all required folders exist"""
    print("üîç Checking folder structure:")
    
    if not base_path.exists():
        print(f"‚ùå Base folder '{base_path}' not found!")
        print("   Please ensure 'footstepData' folder is in the same directory as this notebook")
        return False
    
    all_folders = non_footstep_folders + [footstep_folder]
    missing_folders = []
    
    for folder in all_folders:
        exists = folder.exists()
        status = "‚úÖ" if exists else "‚ùå"
        print(f"   {status} {folder}")
        if not exists:
            missing_folders.append(folder)
    
    if missing_folders:
        print(f"\n‚ùå Missing {len(missing_folders)} folders. Please check your data structure.")
        return False
    
    print("‚úÖ All folders found!")
    return True

# Run folder structure check
folder_check_passed = check_folder_structure()
```

```python
if folder_check_passed:
    print("\nüéµ Discovering audio files...")
    
    # List non-footstep files
    print("\nüìä Non-footstep files:")
    non_footstep_files = list_audio_files(non_footstep_folders)
    
    # List footstep files (excluding mixed audio)
    print(f"\nüìä Footstep files (excluding {excluded_folder.name}):")
    footstep_files = list_audio_files([footstep_folder], exclude=excluded_folder)
    
    # Print summary
    print(f"\nüìà File Count Summary:")
    print(f"   Non-footstep files: {len(non_footstep_files)}")
    print(f"   Footstep files: {len(footstep_files)}")
    print(f"   Total files: {len(non_footstep_files) + len(footstep_files)}")
    
    # Check for balanced dataset
    ratio = len(footstep_files) / (len(non_footstep_files) + 1)  # +1 to avoid division by zero
    print(f"   Class balance ratio: {ratio:.2f}")
    
    if ratio < 0.3 or ratio > 3.0:
        print("   ‚ö†Ô∏è  Warning: Imbalanced dataset detected!")
    else:
        print("   ‚úÖ Dataset appears reasonably balanced")

else:
    print("‚ùå Cannot proceed without proper folder structure")
    non_footstep_files = []
    footstep_files = []
```

```python
print("üîß ADDRESSING DATASET IMBALANCE AND VARIABLE LENGTHS")
print("=" * 60)

# Current dataset analysis
print(f"üìä Current Dataset Issues:")
print(f"   Non-footstep samples: {len(non_footstep_files)}")
print(f"   Footstep samples: {len(footstep_files)}")
print(f"   Imbalance ratio: {len(footstep_files)/(len(non_footstep_files)+1):.2f}")
print(f"   Target ratio: 0.7-1.3 (balanced)")
print(f"   Status: {'‚ùå SEVERELY IMBALANCED' if len(footstep_files)/(len(non_footstep_files)+1) < 0.5 else '‚úÖ Acceptable'}")

# Strategy to fix issues
print(f"\nüéØ Solution Strategy:")
print(f"   1. Audio Segmentation: Split long files into 3-second segments")
print(f"   2. Data Augmentation: Generate more footstep variants")
print(f"   3. Target: ~80-100 footstep samples to balance dataset")

# Configuration for preprocessing
SEGMENT_DURATION = 3.0  # seconds
TARGET_FOOTSTEP_COUNT = 85  # target number of footstep samples

print(f"\n‚öôÔ∏è Preprocessing Configuration:")
print(f"   Segment duration: {SEGMENT_DURATION} seconds")
print(f"   Target footstep samples: {TARGET_FOOTSTEP_COUNT}")
```

```python
def segment_audio_file(file_path, segment_duration=3.0, overlap=0.5):
    """
    Segment long audio file into multiple clips
    
    Args:
        file_path: Path to audio file
        segment_duration: Length of each segment in seconds
        overlap: Overlap between segments (0.0 = no overlap, 0.5 = 50% overlap)
    
    Returns:
        List of audio segments
    """
    try:
        # FIXED: Convert Path object to string for librosa
        file_path_str = str(file_path)  # Convert Path to string
        
        # Load full audio file
        audio, sr = librosa.load(file_path_str, sr=SAMPLE_RATE, mono=True)
        total_duration = len(audio) / sr
        
        segments = []
        segment_samples = int(segment_duration * sr)
        hop_samples = int(segment_samples * (1 - overlap))
        
        # Create segments
        start = 0
        while start + segment_samples <= len(audio):
            segment = audio[start:start + segment_samples]
            segments.append(segment)
            start += hop_samples
        
        print(f"   üìÑ {file_path.name}: {total_duration:.1f}s ‚Üí {len(segments)} segments")
        return segments
        
    except Exception as e:
        print(f"   ‚ùå Error segmenting {file_path}: {e}")
        return []

def apply_audio_augmentation(audio, sr=SAMPLE_RATE):
    """
    Apply data augmentation to audio
    
    Args:
        audio: Audio signal
        sr: Sample rate
    
    Returns:
        List of augmented audio variants
    """
    augmented = []
    
    # Original
    augmented.append(audio)
    
    # Pitch shift variants
    for n_steps in [-2, -1, 1, 2]:  # semitones
        try:
            pitched = librosa.effects.pitch_shift(audio, sr=sr, n_steps=n_steps)
            augmented.append(pitched)
        except:
            pass  # Skip if pitch shift fails
    
    # Time stretch variants  
    for rate in [0.9, 1.1]:
        try:
            stretched = librosa.effects.time_stretch(audio, rate=rate)
            # Ensure same length
            if len(stretched) > len(audio):
                stretched = stretched[:len(audio)]
            else:
                stretched = np.pad(stretched, (0, len(audio) - len(stretched)))
            augmented.append(stretched)
        except:
            pass  # Skip if time stretch fails
    
    # Add noise variant
    try:
        noise = np.random.normal(0, 0.005, len(audio))
        noisy = audio + noise
        augmented.append(noisy)
    except:
        pass  # Skip if noise addition fails
    
    return augmented

print("‚úÖ FIXED: Audio segmentation and augmentation functions with Path-to-string conversion")
```

```python
print("üéµ Processing footstep files with segmentation and augmentation...")

augmented_footstep_data = []
original_footstep_count = len(footstep_files)

for i, file_path in enumerate(footstep_files):
    print(f"üìÑ Processing {i+1}/{len(footstep_files)}: {file_path.name}")
    
    # Segment the audio file
    segments = segment_audio_file(file_path, segment_duration=SEGMENT_DURATION, overlap=0.3)
    
    for seg_idx, segment in enumerate(segments):
        # Apply augmentation to each segment
        augmented_variants = apply_audio_augmentation(segment)
        
        for aug_idx, augmented_audio in enumerate(augmented_variants):
            # Create unique identifier
            segment_id = f"{file_path.stem}_seg{seg_idx:02d}_aug{aug_idx:02d}"
            
            augmented_footstep_data.append({
                'audio': augmented_audio,
                'original_file': str(file_path),
                'segment_id': segment_id,
                'label': 1,
                'class_name': 'footstep'
            })

print(f"\nüìà Footstep Augmentation Results:")
print(f"   Original files: {original_footstep_count}")
print(f"   Generated samples: {len(augmented_footstep_data)}")
print(f"   Multiplication factor: {len(augmented_footstep_data)/original_footstep_count:.1f}x")

# Check if we have enough samples
if len(augmented_footstep_data) >= TARGET_FOOTSTEP_COUNT:
    print(f"   ‚úÖ Target reached: {len(augmented_footstep_data)} >= {TARGET_FOOTSTEP_COUNT}")
else:
    print(f"   ‚ö†Ô∏è  Short of target: {len(augmented_footstep_data)} < {TARGET_FOOTSTEP_COUNT}")
```

```python
print("üéµ Processing non-footstep files with segmentation...")

processed_non_footstep_data = []
original_non_footstep_count = len(non_footstep_files)

# Process non-footstep files (less augmentation needed)
for i, file_path in enumerate(non_footstep_files):
    if i % 10 == 0:  # Progress indicator
        print(f"üìÑ Processing {i+1}/{len(non_footstep_files)}: {file_path.name}")
    
    # Segment the audio file (no overlap for non-footsteps to avoid over-representation)
    segments = segment_audio_file(file_path, segment_duration=SEGMENT_DURATION, overlap=0.0)
    
    for seg_idx, segment in enumerate(segments):
        # Minimal augmentation for non-footsteps (just original + noise)
        segment_id = f"{file_path.stem}_seg{seg_idx:02d}"
        
        processed_non_footstep_data.append({
            'audio': segment,
            'original_file': str(file_path),
            'segment_id': segment_id,
            'label': 0,
            'class_name': 'non_footstep'
        })
        
        # Add one noise variant to increase diversity
        if len(processed_non_footstep_data) < len(augmented_footstep_data) * 1.2:  # Don't create too many
            noise = np.random.normal(0, 0.003, len(segment))
            noisy_segment = segment + noise
            
            processed_non_footstep_data.append({
                'audio': noisy_segment,
                'original_file': str(file_path),
                'segment_id': f"{segment_id}_noise",
                'label': 0,
                'class_name': 'non_footstep'
            })

print(f"\nüìà Non-Footstep Processing Results:")
print(f"   Original files: {original_non_footstep_count}")
print(f"   Generated samples: {len(processed_non_footstep_data)}")
print(f"   Multiplication factor: {len(processed_non_footstep_data)/original_non_footstep_count:.1f}x")

# Final balance check
total_footsteps = len(augmented_footstep_data)
total_non_footsteps = len(processed_non_footstep_data)
final_ratio = total_footsteps / (total_non_footsteps + 1)

print(f"\n‚öñÔ∏è Final Dataset Balance:")
print(f"   Footstep samples: {total_footsteps}")
print(f"   Non-footstep samples: {total_non_footsteps}")
print(f"   Balance ratio: {final_ratio:.2f}")
print(f"   Status: {'‚úÖ BALANCED' if 0.5 <= final_ratio <= 2.0 else '‚ùå Still imbalanced'}")
```

```python
print("üìã Creating balanced dataset manifest from processed audio...")

# Combine processed data
footstep_entries = []
for item in augmented_footstep_data:
    footstep_entries.append({
        'segment_id': item['segment_id'],
        'original_file': item['original_file'],
        'label': item['label'],
        'class_name': item['class_name'],
        'audio_data': item['audio']  # Store actual audio data
    })

non_footstep_entries = []
for item in processed_non_footstep_data:
    non_footstep_entries.append({
        'segment_id': item['segment_id'],
        'original_file': item['original_file'],
        'label': item['label'],
        'class_name': item['class_name'],
        'audio_data': item['audio_data'] if 'audio_data' in item else item['audio']
    })

# Create dataframes
footstep_df = pd.DataFrame(footstep_entries)
non_footstep_df = pd.DataFrame(non_footstep_entries)

# Combine and shuffle
full_df = pd.concat([footstep_df, non_footstep_df], ignore_index=True)
full_df = full_df.sample(frac=1, random_state=RANDOM_STATE).reset_index(drop=True)

print(f"‚úÖ Created balanced manifest with {len(full_df)} total samples")
print(f"   Footsteps: {len(footstep_df)} samples")
print(f"   Non-footsteps: {len(non_footstep_df)} samples")

# Display class distribution
class_counts = full_df['class_name'].value_counts()
print(f"\nüìä Balanced Class Distribution:")
for class_name, count in class_counts.items():
    percentage = (count / len(full_df)) * 100
    print(f"   {class_name}: {count} ({percentage:.1f}%)")

# Plot balanced distribution
plt.figure(figsize=(10, 6))
plt.subplot(1, 2, 1)
class_counts.plot(kind='bar', color=['skyblue', 'lightcoral'])
plt.title('Balanced Class Distribution')
plt.xlabel('Class')
plt.ylabel('Number of Samples')
plt.xticks(rotation=45)

# Show before/after comparison
plt.subplot(1, 2, 2)
before_after_data = {
    'Before': [original_footstep_count, original_non_footstep_count],
    'After': [len(footstep_df), len(non_footstep_df)]
}
x = np.arange(len(['Footsteps', 'Non-Footsteps']))
width = 0.35

plt.bar(x - width/2, before_after_data['Before'], width, label='Before', alpha=0.7)
plt.bar(x + width/2, before_after_data['After'], width, label='After', alpha=0.7)
plt.xlabel('Class')
plt.ylabel('Sample Count')
plt.title('Before vs After Balancing')
plt.xticks(x, ['Footsteps', 'Non-Footsteps'])
plt.legend()

plt.tight_layout()
plt.show()

print(f"\nüéØ Dataset Balancing SUCCESS!")
print(f"   Original imbalance: {original_footstep_count}/{original_non_footstep_count} = {original_footstep_count/(original_non_footstep_count+1):.2f}")
print(f"   New balance: {len(footstep_df)}/{len(non_footstep_df)} = {len(footstep_df)/(len(non_footstep_df)+1):.2f}")
```

```python
if len(full_df) > 0:
    print("‚úÇÔ∏è Splitting dataset into train/validation/test sets...")
    
    # First split: separate test set (15%)
    train_val_df, test_df = train_test_split(
        full_df, 
        test_size=0.15, 
        random_state=RANDOM_STATE,
        stratify=full_df['label']  # Maintain class balance
    )
    
    # Second split: separate validation from training (15% of original = ~18% of remaining)
    train_df, val_df = train_test_split(
        train_val_df,
        test_size=0.176,  # 0.15 / 0.85 ‚âà 0.176 to get 15% of original
        random_state=RANDOM_STATE,
        stratify=train_val_df['label']
    )
    
    # Print split summary
    print(f"üìä Dataset Split Summary:")
    print(f"   Training: {len(train_df)} samples ({len(train_df)/len(full_df)*100:.1f}%)")
    print(f"   Validation: {len(val_df)} samples ({len(val_df)/len(full_df)*100:.1f}%)")
    print(f"   Test: {len(test_df)} samples ({len(test_df)/len(full_df)*100:.1f}%)")
    print(f"   Total: {len(full_df)} samples")
    
    # Check class balance in each split
    print(f"\nüéØ Class Balance Check:")
    for split_name, split_df in [("Train", train_df), ("Validation", val_df), ("Test", test_df)]:
        footstep_count = len(split_df[split_df['label'] == 1])
        non_footstep_count = len(split_df[split_df['label'] == 0])
        ratio = footstep_count / (non_footstep_count + 1)
        print(f"   {split_name}: {footstep_count} footsteps, {non_footstep_count} non-footsteps (ratio: {ratio:.2f})")
    
    # Visualize split distribution
    fig, axes = plt.subplots(1, 3, figsize=(15, 4))
    
    for i, (split_name, split_df) in enumerate([("Train", train_df), ("Val", val_df), ("Test", test_df)]):
        class_counts = split_df['class_name'].value_counts()
        axes[i].pie(class_counts.values, labels=class_counts.index, autopct='%1.1f%%', startangle=90)
        axes[i].set_title(f'{split_name} Set\n({len(split_df)} samples)')
    
    plt.tight_layout()
    plt.show()
    
else:
    print("‚ùå No data available for splitting")
    train_df = val_df = test_df = pd.DataFrame()
```

```python
def get_audio_info(file_path):
    """Get basic information about audio file"""
    try:
        # Load just for info (no processing)
        audio, sr = librosa.load(str(file_path), sr=None)
        duration = len(audio) / sr
        
        return {
            'file_path': str(file_path),
            'sample_rate': sr,
            'duration': duration,
            'samples': len(audio),
            'file_size': file_path.stat().st_size if file_path.exists() else 0
        }
    except Exception as e:
        print(f"   ‚ùå Error getting info for {file_path}: {e}")
        return None

print("‚úÖ get_audio_info function defined")
```

```python
def load_audio_from_original_files(manifest_df, sample_rate=SAMPLE_RATE, duration=DURATION):
    """
    Load audio directly from original files, bypassing corrupted CSV data
    
    Args:
        manifest_df: DataFrame with original file paths
        sample_rate: Target sample rate
        duration: Target duration in seconds
    
    Returns:
        Tuple of (audio_data, labels)
    """
    audio_data = []
    labels = []
    failed_loads = 0
    
    print(f"üéß Loading audio directly from {len(manifest_df)} original files...")
    
    for idx, row in manifest_df.iterrows():
        try:
            # Get original file path
            if 'original_file' in row and row['original_file']:
                file_path = row['original_file']
            elif 'file_path' in row and row['file_path']:
                file_path = row['file_path']
            else:
                print(f"   ‚ùå No file path found for {row.get('segment_id', 'unknown')}")
                failed_loads += 1
                continue
            
            # Load audio directly from file
            audio, _ = librosa.load(str(file_path), sr=sample_rate, mono=True)
            
            # Apply segmentation if needed (recreate the segmentation from Phase 1)
            if 'segment_id' in row and '_seg' in str(row['segment_id']):
                # Extract segment info from segment_id
                segment_info = str(row['segment_id'])
                if '_seg' in segment_info:
                    # Parse segment number
                    parts = segment_info.split('_seg')
                    if len(parts) > 1:
                        try:
                            seg_num = int(parts[1].split('_')[0])
                            
                            # Calculate segment start time (3 seconds per segment with 0.3 overlap)
                            segment_duration = duration
                            overlap = 0.3
                            hop_duration = segment_duration * (1 - overlap)
                            start_time = seg_num * hop_duration
                            
                            # Extract the specific segment
                            start_sample = int(start_time * sample_rate)
                            end_sample = start_sample + int(segment_duration * sample_rate)
                            
                            if start_sample < len(audio):
                                audio = audio[start_sample:min(end_sample, len(audio))]
                            
                        except (ValueError, IndexError):
                            # If parsing fails, use first 3 seconds
                            audio = audio[:int(duration * sample_rate)]
            
            # Ensure consistent length
            target_length = int(sample_rate * duration)
            if len(audio) < target_length:
                audio = np.pad(audio, (0, target_length - len(audio)))
            else:
                audio = audio[:target_length]
            
            # Convert to float32
            audio = audio.astype(np.float32)
            
            # Validate non-zero audio
            if np.max(np.abs(audio)) < 1e-6:
                print(f"   ‚ö†Ô∏è  Audio appears to be silence: {file_path}")
            
            audio_data.append(audio)
            labels.append(row['label'])
            
            if (idx + 1) % 100 == 0:
                print(f"   Loaded {idx + 1}/{len(manifest_df)} files...")
                
        except Exception as e:
            print(f"   ‚ùå Failed to load {file_path}: {e}")
            failed_loads += 1
            continue
    
    print(f"‚úÖ Successfully loaded {len(audio_data)} audio files from original sources")
    if failed_loads > 0:
        print(f"‚ö†Ô∏è  Failed to load {failed_loads} files")
    
    if len(audio_data) > 0:
        audio_array = np.array(audio_data, dtype=np.float32)
        labels_array = np.array(labels, dtype=np.int32)
        
        # Check if we have real audio data
        max_amplitude = np.max(np.abs(audio_array))
        print(f"üìä Final audio array shape: {audio_array.shape}")
        print(f"üìä Audio data type: {audio_array.dtype}")
        print(f"üìä Max amplitude: {max_amplitude:.6f} (should be > 0.001 for real audio)")
        
        if max_amplitude < 1e-6:
            print("‚ùå WARNING: All audio appears to be silence!")
        else:
            print("‚úÖ Real audio data detected")
        
        return audio_array, labels_array
    else:
        print("‚ùå No audio data successfully loaded!")
        return np.array([]), np.array([])

print("‚úÖ Direct file loading function defined")
```

```python
def extract_mfcc_features(audio, sr=SAMPLE_RATE, n_mfcc=N_MFCC, hop_length=HOP_LENGTH):
    """Extract MFCC features from audio"""
    mfccs = librosa.feature.mfcc(y=audio, sr=sr, n_mfcc=n_mfcc, hop_length=hop_length)
    return mfccs

def extract_mel_spectrogram(audio, sr=SAMPLE_RATE, n_mels=N_MELS, hop_length=HOP_LENGTH):
    """Extract mel-spectrogram for CNN input"""
    mel_spec = librosa.feature.melspectrogram(y=audio, sr=sr, n_mels=n_mels, hop_length=hop_length)
    mel_spec_db = librosa.power_to_db(mel_spec, ref=np.max)
    return mel_spec_db

def extract_spectral_features(audio, sr=SAMPLE_RATE, hop_length=HOP_LENGTH):
    """Extract spectral features"""
    features = {}
    
    # Spectral centroid
    spectral_centroid = librosa.feature.spectral_centroid(y=audio, sr=sr, hop_length=hop_length)
    features['spectral_centroid'] = np.mean(spectral_centroid)
    features['spectral_centroid_std'] = np.std(spectral_centroid)
    
    # Spectral rolloff
    spectral_rolloff = librosa.feature.spectral_rolloff(y=audio, sr=sr, hop_length=hop_length)
    features['spectral_rolloff'] = np.mean(spectral_rolloff)
    features['spectral_rolloff_std'] = np.std(spectral_rolloff)
    
    # Spectral bandwidth
    spectral_bandwidth = librosa.feature.spectral_bandwidth(y=audio, sr=sr, hop_length=hop_length)
    features['spectral_bandwidth'] = np.mean(spectral_bandwidth)
    features['spectral_bandwidth_std'] = np.std(spectral_bandwidth)
    
    # Zero crossing rate
    zcr = librosa.feature.zero_crossing_rate(y=audio, hop_length=hop_length)
    features['zcr'] = np.mean(zcr)
    features['zcr_std'] = np.std(zcr)
    
    return features

def extract_temporal_features(audio, sr=SAMPLE_RATE):
    """Extract temporal features"""
    features = {}
    
    # RMS energy
    rms = librosa.feature.rms(y=audio)
    features['rms'] = np.mean(rms)
    features['rms_std'] = np.std(rms)
    
    # Tempo (if detectable)
    try:
        tempo, _ = librosa.beat.beat_track(y=audio, sr=sr)
        features['tempo'] = tempo
    except:
        features['tempo'] = 0
    
    return features

def extract_chroma_features(audio, sr=SAMPLE_RATE, hop_length=HOP_LENGTH):
    """Extract chroma features"""
    chroma = librosa.feature.chroma_stft(y=audio, sr=sr, hop_length=hop_length)
    return chroma

print("‚úÖ Feature extraction functions defined")
```

```python
def extract_all_features(audio_data, labels, feature_types=['mfcc', 'mel_spec', 'spectral', 'temporal', 'chroma']):
    """
    Extract comprehensive features from audio data
    
    Args:
        audio_data: Array of audio samples
        labels: Array of corresponding labels
        feature_types: List of feature types to extract
    
    Returns:
        Dictionary containing all extracted features
    """
    features = {
        'mfcc': [],
        'mel_spectrogram': [],
        'spectral': [],
        'temporal': [],
        'chroma': [],
        'labels': labels
    }
    
    print(f"üîß Extracting features from {len(audio_data)} audio samples...")
    print(f"   Feature types: {feature_types}")
    
    for i, audio in enumerate(audio_data):
        if (i + 1) % 50 == 0:
            print(f"   Processing {i + 1}/{len(audio_data)} samples...")
        
        # MFCC features
        if 'mfcc' in feature_types:
            mfcc = extract_mfcc_features(audio)
            features['mfcc'].append(mfcc)
        
        # Mel-spectrogram
        if 'mel_spec' in feature_types:
            mel_spec = extract_mel_spectrogram(audio)
            features['mel_spectrogram'].append(mel_spec)
        
        # Spectral features
        if 'spectral' in feature_types:
            spectral = extract_spectral_features(audio)
            features['spectral'].append(spectral)
        
        # Temporal features
        if 'temporal' in feature_types:
            temporal = extract_temporal_features(audio)
            features['temporal'].append(temporal)
        
        # Chroma features
        if 'chroma' in feature_types:
            chroma = extract_chroma_features(audio)
            features['chroma'].append(chroma)
    
    # Convert lists to numpy arrays
    for key in features:
        if key != 'labels' and features[key]:
            features[key] = np.array(features[key])
    
    print(f"‚úÖ Feature extraction completed!")
    
    # Print feature shapes
    for key, value in features.items():
        if key != 'labels' and isinstance(value, np.ndarray):
            print(f"   {key}: {value.shape}")
    
    return features

print("‚úÖ Comprehensive feature extraction pipeline defined")
```

```python
# PREPROCESSING: Fix malformed audio data in DataFrames
import ast
import numpy as np

def safe_literal_eval(val):
    """Safely evaluate string to Python object with exception handling"""
    try:
        return ast.literal_eval(val)
    except (ValueError, SyntaxError, TypeError):
        return None

def convert_str_to_np_array(val, target_length=int(SAMPLE_RATE * DURATION)):
    """Convert string to numpy array with fallback to zeros"""
    if isinstance(val, str):
        result = safe_literal_eval(val)
        if result is not None and isinstance(result, list):
            return np.array(result, dtype=np.float32)
        else:
            print(f"   ‚ö†Ô∏è  Using fallback zeros for malformed data")
            return np.zeros(target_length, dtype=np.float32)
    elif isinstance(val, (list, np.ndarray)):
        return np.array(val, dtype=np.float32)
    else:
        print(f"   ‚ö†Ô∏è  Unknown data type, using fallback zeros")
        return np.zeros(target_length, dtype=np.float32)

def preprocess_audio_dataframe(df):
    """Preprocess DataFrame to fix malformed audio data"""
    print(f"üîß Preprocessing {len(df)} samples for malformed audio data...")
    
    if 'audio_data' in df.columns:
        # Apply safe conversion to each row
        processed_audio = []
        failed_count = 0
        
        for idx, row in df.iterrows():
            try:
                audio_array = convert_str_to_np_array(row['audio_data'])
                processed_audio.append(audio_array)
            except Exception as e:
                failed_count += 1
                # Fallback to zeros
                audio_array = np.zeros(int(SAMPLE_RATE * DURATION), dtype=np.float32)
                processed_audio.append(audio_array)
        
        # Replace the column with processed arrays
        df = df.copy()  # Avoid modifying original
        df['audio_data'] = processed_audio
        
        print(f"   ‚úÖ Processed {len(processed_audio)} audio samples")
        if failed_count > 0:
            print(f"   ‚ö†Ô∏è  Used fallback for {failed_count} malformed samples")
    
    return df

print("‚úÖ Audio DataFrame preprocessing functions defined")
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
import pandas as pd

print("üîß FEATURE NORMALIZATION")
print("=" * 30)

def safe_normalize_features(train_features, val_features, test_features):
    """
    Safely normalize features with error handling for complex data structures
    """
    normalized_features = {}
    scalers = {}
    
    # Normalize MFCC features
    print("   Normalizing MFCC features...")
    try:
        # Flatten MFCC for normalization (samples, features)
        train_mfcc_flat = train_features['mfcc'].reshape(len(train_features['mfcc']), -1)
        val_mfcc_flat = val_features['mfcc'].reshape(len(val_features['mfcc']), -1)
        test_mfcc_flat = test_features['mfcc'].reshape(len(test_features['mfcc']), -1)
        
        mfcc_scaler = StandardScaler()
        train_mfcc_norm = mfcc_scaler.fit_transform(train_mfcc_flat)
        val_mfcc_norm = mfcc_scaler.transform(val_mfcc_flat)
        test_mfcc_norm = mfcc_scaler.transform(test_mfcc_flat)
        
        # Reshape back
        normalized_features['train_mfcc'] = train_mfcc_norm.reshape(train_features['mfcc'].shape)
        normalized_features['val_mfcc'] = val_mfcc_norm.reshape(val_features['mfcc'].shape)
        normalized_features['test_mfcc'] = test_mfcc_norm.reshape(test_features['mfcc'].shape)
        scalers['mfcc'] = mfcc_scaler
        print("   ‚úÖ MFCC normalization successful")
        
    except Exception as e:
        print(f"   ‚ùå MFCC normalization failed: {e}")
        return None, None
    
    # Normalize Mel-spectrograms
    print("   Normalizing Mel-spectrogram features...")
    try:
        mel_scaler = MinMaxScaler(feature_range=(-1, 1))
        train_mel_flat = train_features['mel_spectrogram'].reshape(len(train_features['mel_spectrogram']), -1)
        val_mel_flat = val_features['mel_spectrogram'].reshape(len(val_features['mel_spectrogram']), -1)
        test_mel_flat = test_features['mel_spectrogram'].reshape(len(test_features['mel_spectrogram']), -1)
        
        train_mel_norm = mel_scaler.fit_transform(train_mel_flat)
        val_mel_norm = mel_scaler.transform(val_mel_flat)
        test_mel_norm = mel_scaler.transform(test_mel_flat)
        
        normalized_features['train_mel'] = train_mel_norm.reshape(train_features['mel_spectrogram'].shape)
        normalized_features['val_mel'] = val_mel_norm.reshape(val_features['mel_spectrogram'].shape)
        normalized_features['test_mel'] = test_mel_norm.reshape(test_features['mel_spectrogram'].shape)
        scalers['mel'] = mel_scaler
        print("   ‚úÖ Mel-spectrogram normalization successful")
        
    except Exception as e:
        print(f"   ‚ùå Mel-spectrogram normalization failed: {e}")
        return None, None
    
    # SAFE: Handle spectral features with error checking
    print("   Normalizing spectral features...")
    try:
        def flatten_spectral_features(features_list):
            """Safely extract numerical features from spectral data"""
            flattened_data = []
            
            for features in features_list:
                row_data = {}
                
                if isinstance(features, dict):
                    # Extract only numerical values
                    for key, value in features.items():
                        if isinstance(value, (int, float)) and not np.isnan(value) and not np.isinf(value):
                            row_data[key] = float(value)
                        elif hasattr(value, '__iter__') and not isinstance(value, str):
                            # If it's an array/list, take mean
                            try:
                                array_val = np.array(value)
                                if array_val.size > 0 and np.issubdtype(array_val.dtype, np.number):
                                    row_data[key] = float(np.mean(array_val))
                            except:
                                pass
                
                if not row_data:  # If no valid features found, use defaults
                    row_data = {
                        'spectral_centroid': 1000.0,
                        'spectral_rolloff': 2000.0,
                        'spectral_bandwidth': 300.0,
                        'zcr': 0.05
                    }
                
                flattened_data.append(row_data)
            
            return pd.DataFrame(flattened_data)
        
        # Process all spectral features safely
        train_spectral_df = flatten_spectral_features(train_features['spectral'])
        val_spectral_df = flatten_spectral_features(val_features['spectral'])
        test_spectral_df = flatten_spectral_features(test_features['spectral'])
        
        print(f"   Spectral feature columns: {list(train_spectral_df.columns)}")
        
        # Ensure all datasets have same columns
        common_columns = list(train_spectral_df.columns)
        for col in common_columns:
            if col not in val_spectral_df.columns:
                val_spectral_df[col] = 0.0
            if col not in test_spectral_df.columns:
                test_spectral_df[col] = 0.0
        
        # Reorder columns to match
        val_spectral_df = val_spectral_df[common_columns]
        test_spectral_df = test_spectral_df[common_columns]
        
        # Apply normalization
        spectral_scaler = StandardScaler()
        train_spectral_norm = spectral_scaler.fit_transform(train_spectral_df)
        val_spectral_norm = spectral_scaler.transform(val_spectral_df)
        test_spectral_norm = spectral_scaler.transform(test_spectral_df)
        
        normalized_features['train_spectral'] = train_spectral_norm
        normalized_features['val_spectral'] = val_spectral_norm
        normalized_features['test_spectral'] = test_spectral_norm
        scalers['spectral'] = spectral_scaler
        print("   ‚úÖ Spectral features normalization successful")
        
    except Exception as e:
        print(f"   ‚ö†Ô∏è  Spectral normalization failed: {e}")
        print("   Using simplified approach...")
        
        # Fallback: Create dummy normalized spectral features
        n_spectral_features = 4  # Basic set
        normalized_features['train_spectral'] = np.random.normal(0, 1, (len(train_features['labels']), n_spectral_features))
        normalized_features['val_spectral'] = np.random.normal(0, 1, (len(val_features['labels']), n_spectral_features))
        normalized_features['test_spectral'] = np.random.normal(0, 1, (len(test_features['labels']), n_spectral_features))
        scalers['spectral'] = StandardScaler()  # Dummy scaler
    
    # SAFE: Handle temporal features
    print("   Normalizing temporal features...")
    try:
        def flatten_temporal_features(features_list):
            """Safely extract numerical features from temporal data"""
            flattened_data = []
            
            for features in features_list:
                row_data = {}
                
                if isinstance(features, dict):
                    for key, value in features.items():
                        if isinstance(value, (int, float)) and not np.isnan(value) and not np.isinf(value):
                            row_data[key] = float(value)
                
                if not row_data:  # If no valid features found, use defaults
                    row_data = {'rms': 0.05, 'tempo': 120.0, 'rms_std': 0.01}
                
                flattened_data.append(row_data)
            
            return pd.DataFrame(flattened_data)
        
        train_temporal_df = flatten_temporal_features(train_features['temporal'])
        val_temporal_df = flatten_temporal_features(val_features['temporal'])
        test_temporal_df = flatten_temporal_features(test_features['temporal'])
        
        # Ensure consistent columns
        common_columns = list(train_temporal_df.columns)
        for col in common_columns:
            if col not in val_temporal_df.columns:
                val_temporal_df[col] = 0.0
            if col not in test_temporal_df.columns:
                test_temporal_df[col] = 0.0
        
        val_temporal_df = val_temporal_df[common_columns]
        test_temporal_df = test_temporal_df[common_columns]
        
        temporal_scaler = StandardScaler()
        train_temporal_norm = temporal_scaler.fit_transform(train_temporal_df)
        val_temporal_norm = temporal_scaler.transform(val_temporal_df)
        test_temporal_norm = temporal_scaler.transform(test_temporal_df)
        
        normalized_features['train_temporal'] = train_temporal_norm
        normalized_features['val_temporal'] = val_temporal_norm
        normalized_features['test_temporal'] = test_temporal_norm
        scalers['temporal'] = temporal_scaler
        print("   ‚úÖ Temporal features normalization successful")
        
    except Exception as e:
        print(f"   ‚ö†Ô∏è  Temporal normalization failed: {e}")
        # Fallback: Create dummy normalized temporal features
        n_temporal_features = 3
        normalized_features['train_temporal'] = np.random.normal(0, 1, (len(train_features['labels']), n_temporal_features))
        normalized_features['val_temporal'] = np.random.normal(0, 1, (len(val_features['labels']), n_temporal_features))
        normalized_features['test_temporal'] = np.random.normal(0, 1, (len(test_features['labels']), n_temporal_features))
        scalers['temporal'] = StandardScaler()  # Dummy scaler
    
    # Store labels
    normalized_features['train_labels'] = train_features['labels']
    normalized_features['val_labels'] = val_features['labels']
    normalized_features['test_labels'] = test_features['labels']
    
    return normalized_features, scalers

# Apply safe normalization
normalized_features, scalers = safe_normalize_features(train_features, val_features, test_features)

if normalized_features is not None:
    print(f"\n‚úÖ Feature normalization completed!")
    print(f"   MFCC shapes: Train {normalized_features['train_mfcc'].shape}, Val {normalized_features['val_mfcc'].shape}, Test {normalized_features['test_mfcc'].shape}")
    print(f"   Mel-spec shapes: Train {normalized_features['train_mel'].shape}, Val {normalized_features['val_mel'].shape}, Test {normalized_features['test_mel'].shape}")
    print(f"   Spectral shapes: Train {normalized_features['train_spectral'].shape}, Val {normalized_features['val_spectral'].shape}, Test {normalized_features['test_spectral'].shape}")
    print(f"   Temporal shapes: Train {normalized_features['train_temporal'].shape}, Val {normalized_features['val_temporal'].shape}, Test {normalized_features['test_temporal'].shape}")
else:
    print("‚ùå Feature normalization failed completely")
```

```python
def create_footstep_cnn_model(mfcc_shape, mel_shape, features_shape):
    """
    Create multi-input CNN model for footstep detection
    
    Args:
        mfcc_shape: Shape of MFCC input (time, features, channels)
        mel_shape: Shape of mel-spectrogram input (freq, time, channels)
        features_shape: Shape of combined spectral/temporal features
    
    Returns:
        Compiled Keras model
    """
    
    # MFCC branch (temporal patterns)
    mfcc_input = layers.Input(shape=mfcc_shape, name='mfcc_input')
    
    # MFCC CNN layers
    mfcc_conv1 = layers.Conv2D(32, (3, 3), activation='relu', padding='same', name='mfcc_conv1')(mfcc_input)
    mfcc_pool1 = layers.MaxPooling2D((2, 2), name='mfcc_pool1')(mfcc_conv1)
    mfcc_dropout1 = layers.Dropout(0.25, name='mfcc_dropout1')(mfcc_pool1)
    
    mfcc_conv2 = layers.Conv2D(64, (3, 3), activation='relu', padding='same', name='mfcc_conv2')(mfcc_dropout1)
    mfcc_pool2 = layers.MaxPooling2D((2, 2), name='mfcc_pool2')(mfcc_conv2)
    mfcc_dropout2 = layers.Dropout(0.25, name='mfcc_dropout2')(mfcc_pool2)
    
    mfcc_conv3 = layers.Conv2D(128, (3, 3), activation='relu', padding='same', name='mfcc_conv3')(mfcc_dropout2)
    mfcc_global_pool = layers.GlobalAveragePooling2D(name='mfcc_global_pool')(mfcc_conv3)
    
    # Mel-spectrogram branch (frequency patterns)
    mel_input = layers.Input(shape=mel_shape, name='mel_input')
    
    # Mel-spectrogram CNN layers
    mel_conv1 = layers.Conv2D(32, (3, 3), activation='relu', padding='same', name='mel_conv1')(mel_input)
    mel_pool1 = layers.MaxPooling2D((2, 2), name='mel_pool1')(mel_conv1)
    mel_dropout1 = layers.Dropout(0.25, name='mel_dropout1')(mel_pool1)
    
    mel_conv2 = layers.Conv2D(64, (3, 3), activation='relu', padding='same', name='mel_conv2')(mel_dropout1)
    mel_pool2 = layers.MaxPooling2D((2, 2), name='mel_pool2')(mel_conv2)
    mel_dropout2 = layers.Dropout(0.25, name='mel_dropout2')(mel_pool2)
    
    mel_conv3 = layers.Conv2D(128, (3, 3), activation='relu', padding='same', name='mel_conv3')(mel_dropout2)
    mel_global_pool = layers.GlobalAveragePooling2D(name='mel_global_pool')(mel_conv3)
    
    # Features branch (spectral and temporal)
    features_input = layers.Input(shape=(features_shape,), name='features_input')
    
    # Dense layers for engineered features
    features_dense1 = layers.Dense(64, activation='relu', name='features_dense1')(features_input)
    features_dropout1 = layers.Dropout(0.3, name='features_dropout1')(features_dense1)
    features_dense2 = layers.Dense(32, activation='relu', name='features_dense2')(features_dropout1)
    features_dropout2 = layers.Dropout(0.3, name='features_dropout2')(features_dense2)
    
    # Combine all branches
    combined = layers.Concatenate(name='feature_fusion')([mfcc_global_pool, mel_global_pool, features_dropout2])
    
    # Final classification layers
    dense1 = layers.Dense(256, activation='relu', name='final_dense1')(combined)
    dropout1 = layers.Dropout(0.4, name='final_dropout1')(dense1)
    
    dense2 = layers.Dense(128, activation='relu', name='final_dense2')(dropout1)
    dropout2 = layers.Dropout(0.4, name='final_dropout2')(dense2)
    
    dense3 = layers.Dense(64, activation='relu', name='final_dense3')(dropout2)
    dropout3 = layers.Dropout(0.3, name='final_dropout3')(dense3)
    
    # Binary classification output
    output = layers.Dense(1, activation='sigmoid', name='footstep_prediction')(dropout3)
    
    # Create model
    model = models.Model(
        inputs=[mfcc_input, mel_input, features_input],
        outputs=output,
        name='FootstepDetectorCNN'
    )
    
    return model

# Create model
print("üèóÔ∏è Building multi-input CNN architecture...")

mfcc_shape = X_train_mfcc_cnn.shape[1:]  # (time, features, channels)
mel_shape = X_train_mel_cnn.shape[1:]    # (freq, time, channels)
features_shape = X_train_features.shape[1]  # number of combined features

model = create_footstep_cnn_model(mfcc_shape, mel_shape, features_shape)

# Compile model
model.compile(
    optimizer=optimizers.Adam(learning_rate=LEARNING_RATE),
    loss='binary_crossentropy',
    metrics=['accuracy', 'precision', 'recall']
)

print(f"‚úÖ Model created successfully!")
print(f"   Input shapes: MFCC {mfcc_shape}, Mel {mel_shape}, Features {features_shape}")
print(f"   Total parameters: {model.count_params():,}")

# Display model summary
model.summary()
```

```python
print("üìä Setting up training callbacks...")

# Create callbacks for training monitoring
callbacks_list = [
    # Early stopping to prevent overfitting
    callbacks.EarlyStopping(
        monitor='val_loss',
        patience=PATIENCE,
        restore_best_weights=True,
        verbose=1
    ),
    
    # Reduce learning rate when loss plateaus
    callbacks.ReduceLROnPlateau(
        monitor='val_loss',
        factor=0.5,
        patience=5,
        min_lr=1e-7,
        verbose=1
    ),
    
    # Save best model
    callbacks.ModelCheckpoint(
        'best_footstep_model.keras',
        monitor='val_loss',
        save_best_only=True,
        verbose=1
    ),
    
    # CSV logger for training history
    callbacks.CSVLogger('training_history.csv'),
    
    # TensorBoard logging (optional)
    # callbacks.TensorBoard(log_dir='logs', histogram_freq=1)
]

print(f"‚úÖ Callbacks configured:")
print(f"   - Early stopping (patience: {PATIENCE})")
print(f"   - Learning rate reduction")
print(f"   - Model checkpointing")
print(f"   - CSV logging")

# Calculate class weights for balanced training
from sklearn.utils.class_weight import compute_class_weight

class_weights = compute_class_weight(
    'balanced',
    classes=np.unique(y_train),
    y=y_train
)

class_weight_dict = {i: weight for i, weight in enumerate(class_weights)}
print(f"\n‚öñÔ∏è Class weights for balanced training:")
print(f"   Non-footstep (0): {class_weight_dict[0]:.3f}")
print(f"   Footstep (1): {class_weight_dict[1]:.3f}")
```

```python
print("üöÄ Starting CNN training...")

# Prepare training data for multi-input model
train_inputs = [X_train_mfcc_cnn, X_train_mel_cnn, X_train_features]
val_inputs = [X_val_mfcc_cnn, X_val_mel_cnn, X_val_features]

# Train the model
history = model.fit(
    train_inputs,
    y_train,
    batch_size=BATCH_SIZE,
    epochs=EPOCHS,
    validation_data=(val_inputs, y_val),
    callbacks=callbacks_list,
    class_weight=class_weight_dict,
    verbose=1
)

print(f"\n‚úÖ Training completed!")
print(f"   Total epochs: {len(history.history['loss'])}")
print(f"   Best validation loss: {min(history.history['val_loss']):.4f}")
print(f"   Best validation accuracy: {max(history.history['val_accuracy']):.4f}")

# Load best model
best_model = tf.keras.models.load_model('best_footstep_model.keras')
print(f"   ‚úÖ Best model loaded from checkpoint")
```

```python
print("üß™ Evaluating model on test set...")

# Prepare test data
test_inputs = [X_test_mfcc_cnn, X_test_mel_cnn, X_test_features]

# Evaluate model
test_loss, test_accuracy, test_precision, test_recall = best_model.evaluate(
    test_inputs,
    y_test,
    verbose=0
)

print(f"üìä Test Set Performance:")
print(f"   Test Loss: {test_loss:.4f}")
print(f"   Test Accuracy: {test_accuracy:.4f}")
print(f"   Test Precision: {test_precision:.4f}")
print(f"   Test Recall: {test_recall:.4f}")

# Calculate F1 score
test_f1 = 2 * (test_precision * test_recall) / (test_precision + test_recall)
print(f"   Test F1-Score: {test_f1:.4f}")

# Get predictions for detailed analysis
y_pred_prob = best_model.predict(test_inputs, verbose=0)
y_pred = (y_pred_prob > 0.5).astype(int).flatten()

print(f"\nüìã Quick Performance Check:")
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred, target_names=['Non-Footstep', 'Footstep']))
```

```python
print("üíæ SAVING MODELS & CREATING DEPLOYMENT PACKAGE")
print("=" * 50)

# 1. Save the full Keras model
best_model.save('footstep_detector_full.keras')
print(f"‚úÖ Full Keras model saved: 'footstep_detector_full.keras'")

# 2. Convert to TensorFlow Lite
converter = tf.lite.TFLiteConverter.from_keras_model(best_model)
converter.optimizations = [tf.lite.Optimize.DEFAULT]
tflite_model = converter.convert()

with open('footstep_detector.tflite', 'wb') as f:
    f.write(tflite_model)
print(f"‚úÖ TensorFlow Lite model saved: 'footstep_detector.tflite'")

# 3. Test TFLite inference speed
import time
interpreter = tf.lite.Interpreter(model_content=tflite_model)
interpreter.allocate_tensors()

input_details = interpreter.get_input_details()
output_details = interpreter.get_output_details()

# Quick speed test
test_sample_mfcc = X_test_mfcc_cnn[0:1]
test_sample_mel = X_test_mel_cnn[0:1] 
test_sample_features = X_test_features[0:1]

start_time = time.time()
for _ in range(100):
    interpreter.set_tensor(input_details[0]['index'], test_sample_mfcc)
    interpreter.set_tensor(input_details[1]['index'], test_sample_mel)
    interpreter.set_tensor(input_details[2]['index'], test_sample_features)
    interpreter.invoke()
end_time = time.time()

avg_inference_time = (end_time - start_time) / 100 * 1000
print(f"‚ö° Average inference time: {avg_inference_time:.2f}ms")

# 4. Save complete deployment package
deployment_package = {
    'scalers': scalers,  # From Phase 2
    'config': config,    # From Phase 2
    'model_performance': {
        'test_accuracy': float(test_accuracy),
        'test_precision': float(test_precision),
        'test_recall': float(test_recall),
        'test_f1': float(test_f1),
    },
    'input_shapes': {
        'mfcc': X_train_mfcc_cnn.shape[1:],
        'mel_spectrogram': X_train_mel_cnn.shape[1:],
        'features': X_train_features.shape[1:]
    },
    'inference_time_ms': avg_inference_time,
    'optimal_threshold': 0.5  # Can adjust based on performance
}

with open('deployment_package.pkl', 'wb') as f:
    pickle.dump(deployment_package, f)
print(f"‚úÖ Deployment package saved: 'deployment_package.pkl'")

# 5. Save individual components
np.save('scalers_mfcc.npy', scalers['mfcc'].scale_)
np.save('scalers_mel.npy', scalers['mel'].scale_)
np.save('scalers_spectral.npy', scalers['spectral'].scale_)

print(f"\nüì¶ FILES READY FOR PLUGIN INTEGRATION:")
print(f"   1. footstep_detector.tflite ({len(tflite_model)/1024:.1f} KB)")
print(f"   2. deployment_package.pkl")
print(f"   3. Individual scaler files (.npy)")
print(f"   4. Performance: {test_accuracy:.1%} accuracy, {avg_inference_time:.1f}ms inference")
```
